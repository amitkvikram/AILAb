```python
grid1 = grid.copy()
numIter = 100
gamma = 0.99
# actionList = ['left', 'right', 'down', 'up']
actionList = ['right', 'down']
actionToIndex = {'left':(0, -1), 'right':(0, 1), 'down':(1, 0), 'up':(-1, 0)}

def takeAction(i, j, action):
    df = actionToIndex[action]
    nState = (i + df[0], j + df[1])
    if(nState[0] >=0 and nState[0] < 4 and nState[1] >= 0 and nState[1]<4):
        return nState
    return None

def getReward(grid, i, j, action):
    reward = 0
    for a in actionList:
        nState = takeAction(i, j, a)
        if nState is None:
            continue
        if(a == action):
            reward += 0.6*grid[nState]
        else:
            reward += 0.2*grid[nState]
            
    reward += 0.2*grid[i, j]
            
    return reward - 1
        

for _ in range(numIter):
    for i in range(4):
        for j in range(4):
            if(i == 3 and j == 3):
                continue
            maxReward = -1000
            for action in actionList:
                reward = getReward(grid1, i, j, action)
                maxReward = max(reward, maxReward)
            
            grid1[i, j] = maxReward
    
    print(grid1)
    print("\n")
```
